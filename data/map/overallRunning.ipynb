```python
def combine(df_men, df_women) : 
        # Combine the dataframes
    df_combined = pd.concat([df_men, df_women])

    # Group by Nation and sum the medals
    df_total_medals = df_combined.groupby('Nation', as_index=False).sum()

    # Calculate the total medals
    df_total_medals['Total'] = df_total_medals['Gold'] + df_total_medals['Silver'] + df_total_medals['Bronze']

    # Sort by total medals (optional)
    df_total_medals = df_total_medals.sort_values(by='Total', ascending=False).reset_index(drop=True)


    return df_total_medals
```

```python
# Define a function to add suffix to column names
def add_suffix_to_columns(df, suffix):
    renamed_columns = {'Nation': 'Nation'}
    for col in df.columns:
        if col != 'Nation':
            renamed_columns[col] = f'{col}_{suffix}'
    return df.rename(columns=renamed_columns)

# List of DataFrames and corresponding suffixes
dataframes_with_suffixes = [(data0, '1500'), (data1, '100'), (data2, '200'), (data3, '400'), (data4, '800'), (data5, 'full')]

# Reset index for each DataFrame before concatenating
dataframes_with_suffixes = [(df.reset_index(drop=True), suffix) for df, suffix in dataframes_with_suffixes]


# Add suffixes to column names
dataframes_with_suffixes = [(add_suffix_to_columns(df, suffix), suffix) for df, suffix in dataframes_with_suffixes]

# Merge DataFrames on the "Nation" column
merged_data = dataframes_with_suffixes[0][0]  # Start with the first DataFrame
for df, suffix in dataframes_with_suffixes[1:]:
    merged_data = pd.merge(merged_data, df, on='Nation')


merged_data
```
